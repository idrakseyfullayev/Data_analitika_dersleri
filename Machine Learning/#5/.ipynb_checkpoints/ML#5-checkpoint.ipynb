```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('GDP.csv').tail(20)
iller = df[['Year']]
UDM = df['Azerbaijan'].str.replace(',', '').astype(int)

plt.xlabel('Dərs oxuma saatları')
plt.ylabel('Ümumi Daxili Məhsul')
plt.title('1999-2018 Azərbaycan ÜDM')
plt.scatter(iller, UDM, color = 'red', marker = '+')
plt.show()
df
```

```python
req = linear_model.LinearRegression()
req.fit(iller, UDM)
```

```python
plt.xlabel('Dərs oxuma saatları')
plt.ylabel('Ümumi Daxili Məhsul')
plt.title('1999-2018 Azərbaycan ÜDM')
plt.scatter(iller, UDM, color = 'red', marker = '+')
plt.plot(iller, req.predict(iller), color = 'blue')
plt.show()
```

```python
req.predict([[2023]])
```
