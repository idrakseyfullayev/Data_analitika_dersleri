```python
df = pd.read_csv('billionaire.csv')
df['NetWorth'] = df['NetWorth'].str.replace(' B', '')
df['NetWorth'] = df['NetWorth'].str.replace('$', '')
df.head(10)
```

```python
plt.bar(df['Name'].head(10), df['NetWorth'].head(10), color = '#f7e51e')

plt.xlabel('Milyarderler')
plt.ylabel('Pul miqdari')
plt.title('Dunyanin en zenginlerinin servetleri')
plt.show()
```

```python
plt.barh(df['Name'].head(10).iloc[::-1], df['NetWorth'].head(10).iloc[::-1], color = '#f7e51e')

plt.xlabel('Pul miqdari')
plt.ylabel('Milyarderler')
plt.title('Dunyanin en zenginlerinin servetleri')
plt.show()
```
