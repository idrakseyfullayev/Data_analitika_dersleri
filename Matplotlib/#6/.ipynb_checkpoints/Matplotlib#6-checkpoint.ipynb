```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')
filtr = (df['Released_Year'] == 'PG')


illər = df[~filtr]['Released_Year'].astype(int)
il_aralıqları = [1950, 1970, 1990, 2010]

plt.hist(illər, il_aralıqları, edgecolor = 'black')

plt.axvline(illər.median(), color = 'red', label = 'İllərin medianı')
plt.axvline(illər.mean(), color = 'yellow', label = 'İllərin ədədi ortası')

plt.xlabel('İllər')
plt.ylabel('Saylar')
plt.title('Imdb-yə görə ən yaxşı 1000 filmin illər üzrə sayı')

plt.legend()

plt.show()
illər.unique()
```
