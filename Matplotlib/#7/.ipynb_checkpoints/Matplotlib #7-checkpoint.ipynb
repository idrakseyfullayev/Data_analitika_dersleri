```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data.csv')
baxış_sayları = df['views']
bəyənmə_sayları = df['likes']

nisbət = (baxış_sayları/bəyənmə_sayları)

plt.scatter(baxış_sayları, bəyənmə_sayları, c=nisbət, s=bəyənmə_sayları/10000, cmap = 'Oranges'
            , edgecolor = 'yellow', alpha = 0.8)

cbar = plt.colorbar()
cbar.set_label('Baxış saylarının bəyənmə saylarına nisbəti')

#plt.xscale('log')
#plt.yscale('log')

plt.title('TED Talks baxış və bəyənmə sayı qrafiki')
plt.xlabel('Baxış sayları')
plt.ylabel('Bəyənmə sayları')

plt.show()
df
```
