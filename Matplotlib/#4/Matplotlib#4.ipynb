```python
from matplotlib import pyplot as plt
```

```python
oyunlar = [1, 2, 3, 4, 5, 6, 7, 8]

oyunçu1 = [1, 0, 2, 0, 1, 2, 3, 1]
oyunçu2 = [1, 2, 1, 3, 2, 0, 2, 2]
oyunçu3 = [1, 1, 3, 3, 1, 2, 2, 3]
rənglər = ['green', 'red', 'blue']

etiketlər = ['oyunçu 1', 'oyunçu 2', 'oyunçu 3']

plt.stackplot(oyunlar, oyunçu1, oyunçu2, oyunçu3, labels = etiketlər, colors = rənglər)

plt.legend(loc = (0.02, 0.02))
plt.title('3 oyunçunun oyun nəticələri')
plt.show()
```
