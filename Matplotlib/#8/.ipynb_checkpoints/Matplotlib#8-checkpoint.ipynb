```python
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates
```

```python
tarixlər = [
    datetime(2022, 5, 1),
    datetime(2022, 5, 2),
    datetime(2022, 5, 3),
    datetime(2022, 5, 4),
    datetime(2022, 5, 5),
    datetime(2022, 5, 6),
    datetime(2022, 5, 7)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(tarixlər, y, linestyle = 'solid')
plt.gcf().autofmt_xdate()

myFmt = dates.DateFormatter('%d, %b %Y')
plt.gca().xaxis.set_major_formatter(myFmt)

plt.show()
```
