```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('russia_losses_equipment.csv')
günlər = df['day']
tanklar = df['tank']
artileriyalar = df['field artillery']

plt.plot(günlər, tanklar, label = 'Tanklar')
plt.plot(günlər, artileriyalar, label = 'Artileriyalar')


plt.title('Rusiya-Ukrayna müharibəsində Rusiyanın tank və artileriya itkiləri')
plt.xlabel('Günlər')
plt.ylabel('İtkilərin sayı')

plt.fill_between(günlər, tanklar, artileriyalar, where = (tanklar > artileriyalar), color = 'orange', alpha = 0.25, 
                 label = 'Tanklar və artileriyalar arasında olan itkilər')

plt.legend()
plt.show()
df
```
